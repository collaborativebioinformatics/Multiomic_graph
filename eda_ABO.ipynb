```python
# Assuming your DataFrame is named 'df'
# Count the occurrences of each chromosome
chrom_counts = combined['CHROM'].value_counts().sort_index()

# Create a nicer plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=chrom_counts.index, y=chrom_counts.values)

# Add labels and title
plt.xlabel('Chromosome')
plt.ylabel('Number of Entries')
plt.title('Number of Entries per Chromosome')

# Rotate x-axis labels if needed for better readability
plt.xticks(rotation=0)

# Add count values on top of each bar
for i, count in enumerate(chrom_counts.values):
    plt.text(i, count + max(chrom_counts.values)*0.01, str(count), 
             ha='center', va='bottom')

# Improve aesthetics
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
